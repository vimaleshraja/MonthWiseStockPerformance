```python
import pandas as pd

data= pd.read_csv("/content/drive/MyDrive/Equity Data (1 Jan 2020).csv")
pd.DataFrame(data)
```

```python
data.info()
```

```python
data[' DATE1'] = pd.to_datetime(data[' DATE1'])
print(data)
```

```python

data=data.sort_values(["SYMBOL"," DATE1"], axis=0)
#data=data.set_index(["SYMBOL"])

stocks = data[["SYMBOL"," DATE1", " CLOSE_PRICE"]].copy()

pd.DataFrame(stocks.head(530))


```

```python
stocks.info()

stocks["RETURNS"]= stocks.groupby("SYMBOL")[" CLOSE_PRICE"].pct_change(1)
stocks.RETURNS.describe()
stocks["MONTH"]= stocks[' DATE1'].dt.month
stocks['MONTHDATE'] = stocks[' DATE1'].dt.to_period("M")

stocks.RETURNS.describe()
```

```python
stocks.head()
```

```python
stocks['RETURNS1'] = stocks['RETURNS'] + 1
```

```python
import numpy as np

group = stocks.groupby(['SYMBOL', 'MONTHDATE'])
stocks['MONTHRETURN'] = group["RETURNS1"].transform(np.product)
stocks.head()
#stocks.info()
```

```python
stocks['MONTHRETURN'] = stocks['MONTHRETURN']-1
stocks.head()
```

```python
group = stocks.groupby(['SYMBOL', 'MONTHDATE'])
stocks['SD'] = group['RETURNS'].transform(np.std)
stocks.tail()
```

```python
monthlystocks = stocks.drop_duplicates(subset = ['SYMBOL', 'MONTHRETURN'], keep = 'last')
monthlystocks
```

```python
# Top Performing Stocks Monthly Wise

mo= monthlystocks.sort_values(['MONTHDATE', 'MONTHRETURN'], ascending =[True, False])
pd.DataFrame(mo)
```

```python
#Monthlywise Non Top Performer 

mo= monthlystocks.sort_values(['MONTHDATE', 'MONTHRETURN'], ascending =True)
pd.DataFrame(mo)
```
